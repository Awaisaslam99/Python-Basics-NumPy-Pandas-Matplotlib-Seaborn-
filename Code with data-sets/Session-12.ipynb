# Namespaces:

```python
# A namespaces is a space that holds names(identifiers). Programmatically speaking, namespaces are dictionary 
# of identifiers(keys) and their objects(values).
```

4 types of namespaces:

```python
# 1- Builtin namespace
# 2- Global namespace
# 3- Enclosing namespace
# 4- Local namespace
```

Scope & LEGD(Local,Enclosing,Global,Builtin) Rule:

```python
# A scope is a textual region of a python program where a namespace is directly accessible.

# LEGB Rules:
# The interpreter searches for a name from the inside out, 1st looking in the local, then enclosing, global, and finally the 
# builtin scope. If the interpreter does'nt find the name in any of these locations,then python raises a NameError exception. 
```

```python
# Local & global Scope:

a=2 # Global scope, bcoz it's the part of the global code..

def temp():
    b=3         # Local scope, bcoz it's the part of function scope..
    print(b)
    
temp()
print(a)
```

```python
# Local & Global with the same name:

a=2 

def temp():        
    a=3
    print(a)
    
temp()
print(a)
```

```python
# Local does not have, Global has:

a=2 

def temp():        
    print(a)    # Local can access global .. but can't change the global...
    
temp()
print(a)

```

```python
# local & global ->editing global:

a=2 

def temp():        
    a+=1
    print(a)
    
temp()
print(a)

# Local can access global .. but can't change the global...
```

```python
# we can change global by calling global,but it's not recomended..it's not good programming practice....

a=2 

def temp():    
    global a
    a+=1
    print(a)
    
temp()
print(a)
```

```python
# Local can access global, but not edit or change them.. 
```

```python
# local & global-> global creted inside the local...again it's not good programming practice..just for knowladge.. 

def temp():    
    global a
    a=1
    print(a)
    
temp()
print(a)

```

```python
# local & global-> function parameter is local..
def temp(z):    
    
    print(z)
    
a=2
temp(5)
print(a)
```

Enclosing Scope(nonlocal): 

```python
# Enclosing (or nonlocal) scope is a special scope that only exists for nested functions..


def outer():
    a=3               # Enclosing
    def inner():
        print(a)
    inner()
    print('Outer function')
a=1    
outer()
print('Main program')
```

# Decorators:

```python
# A decorator in python is a function that recieve another function as input and add some functionality(decoration) to it 
# and returns it.

# This can happen only bcoz python function are 1st class citizens.
```

```python
# There are two types of decorators avaliabe in python:

# 1- Built-in decorators like @staticmethod, @classmethod, @abstractmethod and #property etc...if you see "@" it means it's a decorator..

# 2- User defined decorators like that we programmers can creat according to our needs...
```

```python
# Simple decorator Example:

def decorator(func):
    def wrapper():     # btw func inside the decorator called wrapper..
        print('*************')
        func()
        print('*************')
    return wrapper

def hello():
    print('Hello')
    
a= decorator(hello)
a()
```

```python
# same example but Better syntax:

def decorator(func):
    
    def wrapper():
        print('*************')
        func()
        print('*************')
    return wrapper

# 1st func always a decorator..
@decorator
def hello():
    print('Hello')
    
hello()  
```

```python
# execution time measurement:

import time

def timer(func):
    def wrapper():
        start= time.time()      # current time..
        func()
        print('Time taken by',func.__name__ ,time.time() - start,'secs')
    return wrapper

@timer
def hello():
    print("Hello world")
    time.sleep(2)
    
@timer
def display():
    print('Displaying time')
    time.sleep(3)
    
hello()
display()
```
