 Write OOP classes to handle following scenarios: 

```python
# 1- User can creat and view 2D coordinates.
# 2- User can find out the distance between 2 coordinates.
# 3- User can find the distance of a coordinate from origin.
# 4- User can check if a point lies on a given line.
# 5- User can find the distance between a given point and a given line.
```

```python
class Point():
    def __init__ (self,x,y):
        self.x_cod=x       # where x_cod , y_cod are variables that are called instance v ariable.
        self.y_cod=y      
        
        # 1- User can creat and view 2D coordinates.
    def __str__(self):
        return '<{},{}>'.format(self.x_cod,self.y_cod)
    
    # 2- User can find out the distance between 2 coordinates.
    def euclidean_distance(self,other):
        return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5
    
    # 3- User can find the distance of a coordinate from origin.
    def distance_from_origin(self):
         return (self.x_cod**2 - self.y_cod**2)**0.5 # or use next on sytax..
        #return self.euclidean_distance(Point(0,0))
        
    # 4- User can check if a point lies on a given line.
class Line():
    def __init__(self,A,B,C):
        self.A=A
        self.B=B
        self.C=C
        
    def __str__(self):
        return '{}x + {}y + {}=0'.format(self.A,self.B,self.C)
    
    # 4- User can check if a point lies on a given line.
    def point_on_line(line,point):
        if line.A *point.x_cod + line.B*point.y_cod + line.C==0:
            return "lies on the line"
        else:
            return 'Does not lies on the line'
        
    # 5- User can find the distance between a given point and a given line.    
    def shortest_distance(line,point):
        return abs(line.A*point.x_cod + line.B*point.y_cod + line.C)/(line.A**2 + line.B**2 )**0.5
```

```python
l1=Line(1,1,-2)
p1=Point(1,1)
print(l1)
print(p1)

l1.shortest_distance(p1)
```

```python
#Example.. ATM code:

class ATM():

    

    def __init__(self):   
        self.pin=''
        self.__balance= 0    #private
# when we do private to avariable then it convert into this syntax(_ATM__balnce) (_ class_name __variable_name)
        self.menu()
        
    def __menu(self):               # private
        user_input= input("""
        Hi! How can i help you ?
        
        1: Enter 1 to create key.
        2: Enter 2 to change pin.
        3: Enter 3 to check balance.
        4: Enter 4 to withdraw.
        5: Enter anything to exit.
    
        """)
        
        if user_input=='1':
             self.creat_pin()
        elif user_input=='2':
             self.change_pin()
        elif user_input=='3':
             self.check_balance()
        elif user_input=='4':
             self.withdraw()
        else:
            exit()
            print('Have a good day!')

    def creat_pin(self):
        user_input= input("Enter your pin")
        self.pin=user_input
        
        user_balance=int(input('Enter balance'))
        self.balance= user_balance
        
        print('pin created successfully!')
        self.menu()
    
    def change_pin(self):
        old_pin= input('Enter old pin')
        
        if old_pin==self.pin:
            new_pin=input('Enter new pin')
            self.pin=new_pin
            print('pin change successful')
            self.menu()
        else:
            print('Nope! thats wrong pin')
            self.menu()
            
    def check_balance(self):
        user_pin=input('Enter your pin')
        if user_pin==self.pin:
            print('Your balance is  ',self.__balance)    # private
            self.menu()
        else:
            print('Get-out')
            self.menu()
            
    def withdraw(self):
        user_pin=input('Enter your pin')
        if user_pin==self.pin:
            amount=int(input('Enter withdraw amount'))
            if amount<=self.__balance:                      # private
                self.balance=self.__balance - amount
                print('Withdrawl successfully. Balance is:  ', self.__balance)          # private
                
            else:
                print("Nikl bsdk")
                
        else:
            print('Get-out')
            
        self.menu() 
    
obj=ATM()
```
