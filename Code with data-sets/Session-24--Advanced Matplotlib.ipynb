# Colored Scatterplot:

```python
iris= pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species']=   iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)


# here we assign species into integers....
```

```python
                                 # parameter 'c' it give color to a dot, and 'cmap' use to change colors... 
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7) 
                                                                                      #'alpha' use for dot visuals...
plt.xlabel('Sepal Lenght')
plt.ylabel('Petal Lenght')
plt.colorbar()  # it represent colors of values...
```

# Annotations:

Using annotation we can place text on every dot..

```python
df= pd.read_csv('batter.csv')
df.head()
```

```python


sample_df= df.head(100).sample(25,random_state=5) 
sample_df
```

```python

plt.figure(figsize=(15,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
# there is 25 batsman in the graph, we can give a name to a every dot..

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
  # if you want to give name to your dot, use 'plt.text()'...you can also change color, size..
plt.text(1,5,'p1',fontdict={'size':15,'color':'brown'})
plt.text(2,6,'p2')
plt.text(3,7,'p3')
plt.text(4,8,'p4')
```

# Horizontal and vertical lines:

```python

plt.figure(figsize=(15,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

        #pass numbers like, if u want to creat H-line use y-axis no, if u want V-line use x-axis no...
plt.axhline(130,color='red')  # use for creat horizental line in the graph....
plt.axhline(140,color='green')  # 140 plus strik_rate
plt.axvline(30,color='red')   # use for creat vertical line in the graph....
          

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
    
    
    
    
# with using H & V lines you can improve your analysis.....
```

You can creat multiple graph using 'Subplots'....

```python


# simple way to creat graph..


#plt.figure(figsize=(15,6))

#plt.scatter(df['avg'],df['strike_rate']) 
#plt.title('somthing')
#plt.xlabel('avg')
#plt.ylabel('strik_rate')
#plt.show()



# another way to creat graphes..


fig, ax= plt.subplots(figsize=(15,6))  # in this graph use this to change graph size..
ax.scatter(df['avg'],df['strike_rate'],color='red',marker='+')


ax.set_title('somthing') # in this graph use this function to creat title..

ax.set_xlabel('Avg')
ax.set_ylabel('Strike_rate')

fig.show()


# whats the benefit of this another syntax to creat graph???
# we can creat multiple graphes by using this function....
```

```python



fig,ax= plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(df['avg'],df['strike_rate'],color='red')

ax[1].scatter(df['avg'],df['runs'])

ax[0].set_title('Avg vs Strike_rate')
ax[0].set_ylabel('strike_rate')


ax[1].set_title('Avg vs runs')
ax[1].set_ylabel('runs')
ax[0].set_ylabel('Avg')

# in output there's overlaping, to solve thius issue use 'sharex' parameter it remove x=axis points of upper graph...
```

```python


fig,ax= plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(df['avg'],df['strike_rate'],color='red')
ax[0,1].scatter(df['avg'],df['runs'])

ax[1,0].hist(df['avg'])
ax[1,1].hist(df['runs'])

```

```python
# there is another way to creat graph....

fig= plt.figure()


ax1=fig.add_subplot(1,2,1)   # "(2,1,1)" it's means 2 rows, 1 col, 1st graph..
ax1.scatter(df['avg'],df['strike_rate'],color='red')

# if you want to make multiple graph, than reapeat..

ax2= fig.add_subplot(2,2,2)
ax2.hist(df['runs'])


ax3= fig.add_subplot(2,2,3)
ax3.hist(df['avg'])
```

# 3D Surface Plots:

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy= np.meshgrid(x,y)


# Meshgrid is used to create rectangular structures from the given arrays, representing the matrix's indexing...
```

```python
z= xx**2 + yy**2   # provide formula to creat graph...
z.shape
```

```python
fig =plt.figure(figsize=(12,8))

ax= plt.subplot(projection='3d')

a =ax.plot_surface(xx,yy,z,cmap='viridis')  # this is the main func you use in 3D surface plots..

fig.colorbar(a)   # give variable in this func to use this func..
```

```python
z= np.sin(xx) + np.cos(yy)
```

```python
fig =plt.figure(figsize=(12,8))

ax= plt.subplot(projection='3d')

a =ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(a)
```

```python
z= xx**2 + yy**2

fig =plt.figure(figsize=(12,8))

ax= plt.subplot(projection='3d')

a =ax.plot_surface(xx,yy,z,cmap='viridis') 

fig.colorbar(a) 

# if you ant to see this graph in 2D plot...use cntour plots...
```

# Contour Plots:

A contour plot represents a 3-D surface by plotting lines that connect points with common z-values along a slice.

```python
z= xx**2 + yy**2

fig =plt.figure(figsize=(12,8))

ax= plt.subplot()

a =ax.contour(xx,yy,z,cmap='viridis')  # main func...

fig.colorbar(a)  
```

```python
# there is one more function 'contourf'  it fill the lines with colors..


z= xx**2 + yy**2
fig =plt.figure(figsize=(12,8))

ax= plt.subplot()

a =ax.contourf(xx,yy,z,cmap='viridis')  

fig.colorbar(a)   


# opprr 3D graph tha, or ye 2d graph ha, isko smjhna kesy ha, listen, oprr waly 3D graph ko hum oprr se dekh rhy ha,
# jo dark colors ha wo ghraii ko show kr rhy ha or light colors hight ko show k rhy ha,,,
```

# Heatmap:

Heatmaps are a way to visualize data using color.... 

```python
delivery= pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df= delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) &  (delivery['batsman_run'] == 6) ] 
```

```python
grid= temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
# to creat 'Heatmap' use 'imshow()' ..


plt.figure(figsize=(12,7))
plt.imshow(grid)   # main func...

plt.yticks(delivery['overs'].unique(), list(range(1,21)))  # it show all overs...
plt.xticks(np.arange(0,6), list(range(1,7)))    # it show all ballnumber..


plt.colorbar()
```
