# Tuples:

```python
# Tuples similer to list, the difference between list & list is that tuples are immutable...
```

```python
#Creating tuples:
t1=()    # empty tuple..
print(t1)

#single item tuple:
t2=(2,)
print(type(t2))  # this is not a tuple,it's string, so to make it tuple use , after item like..

#Homogenous:
t3=(1,2,3)
print(t3)

#Hetrogenous:
t4=(1,2.5,True,'Hello')
print(t4)

#2D tuple:
t5=(1,2,(3,4))
print(t5)

# using type conversation:
t6=tuple('Hello')
print(t6)
```

```python
# How to access item from tuple:

#  1-Indexing:
print(t3)
print(t3[1])     #positive indexing...
print(t3[-1])    #Negative indexing...
```

```python
# 2-Slicing:
print(t3[0:2])
print(t3[::-1])
```

```python
# Edit & Adiing items in tuples:
#tuples are immutable, so we cant't edit or adding items in tuples...
```

```python
# Delete items:
# We can del whole tuple but can't del single or double items in tuple...
del t1
print(t1)
```

```python
# Operations on tuple:
# 1- (+ , *):
t1=(1,2,3,4,5)
t2=(6,7,8,9,10)
print(t1+t2)
print(t1*2)
    
# 2-Membership:    
print(1 in t1)

#  3-Iteration:
for i in t1:
    print(i)
```

```python
# Tuple Functions:

t=(1,2,3,4,5)
print(len(t))
print(sum(t))
print(min(t))
print(max(t))
print(sorted(t))
print(sorted(t,reverse=True))
```

```python
# count:
t.count(5)
```

```python
# index:
t.index(4)
```

# Sets:

```python
# Creating sets:
#Empty set:
#s={}               #This is'nt empty cozz set & dictionary has same syntax, so to creat empty set use type conversian
s=set()
print(s)
print(type(s))

# 1D & 2D set:
s1={1,2,3}
print(s1)                 #There is no 2D set in sets cozz set can't contain mutable data types..


#Homo & Hetrogenous:
s3={1,2.3,True,'Hello'}   # python treat True as a 1. So true can't print cozz there is no duplecate in sets, 
print(s3)               

#Using type conversian:
s4=set([1,2,3,4])
print(s4)

#Duplication not allowed:
s5={1,2,1,3}
print(s5)

#Set can't have mutable itmes:
s6{1,2,[1,2]}
print(s6)      #print error cozz set can't contain mutable data types..

```

```python
# Sets are unordered :
a={1,2,3}
b={3,2,1}
print(a==b)
```

```python
#Accessing items:
#Accessing items does not allow cozz sets are unoredered.
# we can't do indexing & slicing in sets cozz sets are unoredered..
```

```python
#Editing items:
#For editing items we use indexing & slicing.  
#But in sets we can't do indexing & slicing cozz sets are unoredered..
```

```python
#Adding items:

# 1-add:   use to add single item in set...
s={1,2,3,4}
s.add(5)             #we can't choose position cozz sets follow hashing concept..
print(s)

# 2- update:   use to add multiple items in set...
s.update([5,6,7,8])
print(s)
```

```python
#Deleting items:

#1-del:  delete whole set..
s={1,2,3,4,5}
print(s)
del s        #we can't delete particuler item cozz set are unordered..
print(s)             
```

```python
#2-discard:  we delete item by telling number..
s={1,2,3,4,5}
s.discard(5)         # if we throuogh diff item that not exist in set, it print set
print(s)

```

```python
#3-remove:
# remove & discard are same, 
s={1,2,3,4,5}
s.remove(5)   
print(s)
#the diff b/w discard & remove, discard print set if we  throuogh diff item that not exist, but remove print error..
```

```python
#4-pop:    delete random item..
s={1,2,3,4,5}
s.pop()
print(s)
```

```python
# 5-clear:   juts clear set..
s={1,2,3,4,5}
s.clear()
print(s)
```

```python
#union/update:
s1={1,2,3,4,5}
s2={4,5,6,7,8}
s1.union(s2)

s1.update(s2)            # 1st set update permanntly..
print(s1)
print(s2)
```

```python
#Intersection/Intersection_update:
s1={1,2,3,4,5}
s2={4,5,6,7,8}
s1.intersection(s2)

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
#difference/difference_update:
s1={1,2,3,4,5}
s2={4,5,6,7,8}
s1.difference(s2)

s1.difference_update(s2)
print(s1)
print(s2)
```

```python
#Symmetric_difference/Symmetric_difference_update:
s1={1,2,3,4,5}
s2={4,5,6,7,8}
s1.symmetric_difference(s2)

s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
#isdisjoint:
s1={1,2,3,4,5}
s2={4,5,6,7,8}
s1.isdisjoint(s2)
```

```python
#issubset:
s1={1,2,3,4,5}
s2={4,5}
s2.issubset(s1)
```

```python
#issuperset:
s1={1,2,3,4,5}
s2={4,5}
s1.issuperset(s2)
```

```python
# copy:
s1={1,2,3}
s2=s1.copy()
print(s1)
print(s2)
```

Dictionary Comprehension:  
{Key: value for vars in iterable}

```python
#Print 1st 10 no. and their squares:
{i:i**2 for i in range(1,11)}
```

```python
#using existing dict,,:
distance={'lhr':100,'mdk':200,'bwl':300}
{key:value*1.62 for (key,value) in distance.items()}
```

```python
#using zip:
days={'sunday','monday','tuesday','wednesday','Thursday','friday','saturday'}
temp_c={30.5,32.6,31.8,33.4,29.2,31.3,23.9}
{i :j for (i,j) in zip(days,temp_c)}
```

```python
#using if condition:
p={'phone':10,'laptops':12,'charger':9,'tablet':0}


{key:value for(key,value) in p.items() if value>0}
```

```python
#nested comprehension:
# print tables of no. from 2 to 4:
{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```

```python
#Without comprehenssion:
#print tables of no. from 2 to 4:
for i in range(2,5):
    print(f'Table of {i} :')
    for j in range(1,11):
        print({j:i*j})
```
