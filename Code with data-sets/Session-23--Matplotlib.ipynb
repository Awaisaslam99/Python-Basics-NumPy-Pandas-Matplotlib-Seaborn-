# 2D Line plot:

```python
# plot( X , Y ) creates a 2-D line plot of the data in Y versus the corresponding values in X ...
```

```python
# it's use for Bivariate Analysis..
# categorical -> numerical and numerical -> numerical
# Use case - Time series data
```

```python
# ploting a simple function:

price= [48000,54000,57000,49000,47000,45000]
year= [2015,2016,2017,2018,2019,2020]
    
    #1st is x-axis, 2ns is y-axis..
plt.plot(year,price)
```

```python
# from a pandas dataframe:

batsman= pd.read_csv('sharma-kohli.csv')
batsman


plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
## plotting multiple plots


plt.plot(batsman['index'],batsman['V Kohli'])

plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title:

plt.plot(batsman['index'],batsman['V Kohli'])

plt.plot(batsman['index'],batsman['RG Sharma'])
  
    
    
# we acn Also add title...
plt.title('R Sharma VS V Kohli Career Comparison')



plt.xlabel('Season')  # for x=axis label
plt.ylabel('Runs Scored')  # for y-axis label
```

```python
# colors(hex) and line(width and style) and marker(size):

# hex code search on google, it'll give you a every color code you want..

plt.plot(batsman['index'],batsman['V Kohli'],color='#2D670E')  # change line color

plt.plot(batsman['index'],batsman['RG Sharma'],color='#183AE0')
  
    
    
# we acn Also add title...
plt.title('R Sharma VS V Kohli Career Comparison')



plt.xlabel('Season')  # for x=axis label
plt.ylabel('Runs Scored')  # for y-axis label


```

```python
# you can also change linestyle(to much options) and linewidth:

plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# you can change points in line: use 'marker' there's diff style in marker like 'D',"+",'<' etc
# you can also change size of marker: use 'markersize'


plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='D')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

```

```python
# we can also give name to the lines: use 'label', and also use 'plt.legend()' function when you use 'label'..


plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='D',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

```python
# limiting axes:


price= [48000,54000,57000,49000,47000,45000,4500000]
year= [2015,2016,2017,2018,2019,2020,2021]
          
plt.plot(year,price)  # we add big amount in 2021, so just big amount variation dekhy gi,

# if you ant to see all variations, use this function..
plt.ylim(0,100000)  # see all variation comes
plt.xlim(2017,2019) # if you ant see just some years...
```

```python
# we also see grid in graph:n use 'plt.grid()' function...

plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='D',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

# Scatter Plots:

A scatter plot is a diagram where each value in the data set is represented by a dot.     

Bivariate Analysis         
numerical vs numerical      
Use case - Finding correlation

```python
# plt.scatter simple function:

x= np.linspace(-10,10,50)
x

y= 10*x + 3 + np.random.randint(0,300,50)
y

plt.scatter(x,y)

```

```python
# plt.scatter on pandas data:

df= pd.read_csv('batter.csv')
df= df.head(50)
df
```

```python


plt.scatter(df['avg'],df['strike_rate'],color='red',marker='D')
plt.title(['Average and Strike-rate analysis of top 50 batsman'])

plt.xlabel('Average')
plt.ylabel('SR')

```

```python
# if you ant to see k customer ketny logo k saath aaya tha, use 'size(s)' function...

tips= sns.load_dataset('tips')
tips

plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)


```

```python
# scatterplot using plt.plot: pass 'o' parameter...
# why we use plt.plot to creat scatter function, bcozz this is faster technique than scatter function....

plt.plot(tips['total_bill'],tips['tip'],'o')
```

# bar chart:

A type of data visualization used to represent data in the form of rectangular bars.      

Bivariate Analysis       
Numerical vs Categorical      
Use case - Aggregate analysis of group

```python
# simple bar chart:

children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black') # all functions applicable, jo upper prhu ha...
```

```python
# horizontal bar chart:

plt.barh(colors,children,color='black')
```

```python
# color and label:

df= pd.read_csv('batsman_season_record.csv')
df
```

```python


plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]) ,      df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'],width=0.2,color='blue')


#np.arange(df.shape[0]) # run this to see whats happen..

plt.xticks(np.arange(df.shape[0]),df['batsman'])  # use 'xticks' to replace x=axis names..
# you see in ouput there's is some additional ifo, to remove this use 'show'..

plt.show()
```

```python
# a problem:

children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')


# see in output, x-axis name is overlaped...to avoid this use 'xticks' with parameter 'rotation'..
plt.xticks(rotation= 'vertical')



plt.show()
```

```python
# Stacked Bar chart:


  #  batsman total runs..
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')  # 
plt.bar(df['batsman'],df['2015'],bottom=df['2017'] + df['2016'],label='2015') # add upper 'bottom' parameter

plt.legend()
plt.show()
```

# Pie Chart

Pie charts show the size of items (called wedge) in one data series, proportional to the sum of the items.        

Univariate/Bivariate Analysis      
Categorical vs numerical       
Use case - To find contibution on a standard scale

```python
# simple data:

data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']

plt.pie(data, labels=subjects)  # labels use for show labels in pie graph..




plt.show()
```

```python
# dataset:

df= pd.read_csv("gayle-175.csv")
df

plt.pie(df['batsman_runs'], labels=df['batsman'],autopct='%0.1f%%') # 'autopct='%0.1f%%' this parameter show percentage.. 

plt.show()
```

```python
# percentage and colors:

plt.pie(df['batsman_runs'], labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown']) 
plt.show()
```

```python
# explode shadow:

# use 'explode' to explode something, provide value like o.1 etc, and it explode..
plt.pie(df['batsman_runs'], labels=df['batsman'],autopct='%0.1f%%', explode=[0.3,0,0,0,0,0.1],shadow=True) 
                                                                             # 'shadow' give looks to your graph..
plt.show()
```
