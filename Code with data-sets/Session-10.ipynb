# Writing to file:

Case-1--if you ant to write and the file is not present:

```python
# How to creat files:

# 1st file name and then if you want to write use 'w' or want to read use 'r'...
f= open('sample.txt','w')
       # what you want to write in a file...
f.write('Hello world')
   # if you want to close file...
f.close()

# if the file is closed, this will not worked...
# f.write('Hello')
```

Write multiple strings:

```python

f= open('sample.txt1','w')
f.write('Hello world')
f.write('\nHow are you ?')
f.close()
```

Case-2--if the file is already present:

```python


# the problem with w mode
f= open('sample.txt','w')    
# if you want to write on existing file the old content will removed and new content will be added.
f.write('\nAwais Aslam')
f.close()
```

```python


#Introdusing with append mode:
f=open('sample.txt1','a')
# if you want to write new content and does'nt remove old content , so use  'a' instead of 'w'
f.write('\nI am fine, what about you?')
f.close()
```

How exactly open() works??

Write Lines:

```python
L = ['Hello\n', 'Hi\n', 'how are you\n', 'I am fine']    # list
f = open('sample.txt', 'w')
# if you want to write lines use 'writelines' keyword
f.writelines(L)
f.close()
```

# Reading from files:

```python
# How to read from file:
# using read 'r':

f=open('sample.txt','r')
s=f.read()
f.close()
print(s)
```

```python
# if you want to read some characters :

f=open('sample.txt','r')
# add number, how many chars you want to read...
s=f.read(10)
f.close()
print(s)
```

How to read line-by-line:

```python
# Id you want to read by lines:

# using readline:

f=open('sample.txt','r')
print(f.readline(),end='')
print(f.readline(),end='')
f.close()
```

```python
# Read entire data using readline:

f=open('sample.txt','r')

while True:
    data=f.readline()
    if data=='':
        break
    else:
        print(data,end='')
    
f.close()
```

Using Context Manager(With):

```python
# 1- It's a good idea to close a file usage as it will free up the resources.
# 2- If we close it, garbadge collector would close it.
# 3- With keyword closes the file as soon as the usage is over.
```

```python
# using with: 

with open('sample.txt1','w')as f:   
    f.write('Awais Aslam')
    
# using with keyword we don't need to write "f.close()", with keyword automaticaly close file..

#f.write('HAHA')
```

```python
# want to read using with keyword:

with open('sample.txt1','r')as f:
    print(f.read())
```

```python
# if you want to read 10 char and than next 10 char.

with open('sample.txt','r')as f:
    print(f.read(10))
    print(f.read(10))
```

```python
# to load a big file in memory:

big_L=['Hello world' for i in range(100)]

with open('big.txt','w') as f:
    f.writelines(big_L)
```

```python
with open('big.txt','r')as f:
    chunk_size=10
    
    while len(f.read(chunk_size))>0:
        print(f.read(chunk_size),end='***')
        f.read(chunk_size)
```

Functions:

```python
# 1- tell function: tell() method returns current position of file object.

with open('sample.txt','r')as f:
    print(f.read(10))
    print(f.tell())      # tell functions tells us how much char you print and whats the next next char is going to print.
```

```python
# 2- seek function:

with open('sample.txt','r')as f:
    print(f.read(8))
    print(f.tell())
    f.seek(15)     # seek() function is used to change the position of the File to a given specific position.
    print(f.read(10))
    
    print(f.tell())
```

```python
# seek during write:

with open('sample.txt','w')as f:
    f.write('Hello')
    f.seek(0)
    f.write('x')
```

# Problems with working in text mode:

```python
# 1- can't work with binary files like images.
# 2- not good for other data=type like int/float/list/tuples..
```

```python
# working with binary file:
with open('IMG_20230424_142635.jpg','r') as f:
    f.read()      # it print error bcoz image is binary...
# we can't read binary files like this...
```

```python
# to read binary file use rb,wb:

with open('IMG_20230424_142635.jpg','rb')as f:   #rb: read binary
    with open('Picture_copy.jpg','wb') as wf:    #wb: write bianry
        wf.write(f.read())
        
```

```python
# working with other data-types:

with open('sample.txt','w')as f:
    f.write(5)  # it print error bcoz argument is'nt string...
```

```python
# there is only string as argument not other data-type..
with open('sample.txt','w')as f:
    f.write('5')
```

```python
# same with read 
with open('sample.txt','r')as f:
    print(int(f.read()) + 5)  # using type converstion
```

```python
# More complex data:

# we use dict:

d={
    'name':'Awais',
    'age':21,
    'gender':'male'
}
with open('sample.txt','w')as f:
    f.write(str(d))
```

```python
# read dict:

with open('sample.txt','r')as f:
    print(f.read())
    print(type(f.read()))
```

# Seriallization & Deseriallization:

```python
# Seriallization: process of converting python data typs to JSON format.
# Deseriallization: Process of converting JSON to python data types.

#JSON(JavaScript Object Notation): JSON is a universal text-format, and understood by all programming languages.
```

```python
# Seriallization using json module:
# list
import json
L=[1,2,3,4]
with open('demo.json','w')as f:
    json.dump(L,f)
```

```python
# dict:
d={
    'name':'Awais',
    'age':21,
    'gender':'male'
}
with open('demo.json','w')as f:
    
    json.dump(d,f,indent=4) # use indent="space" for better representation
```

```python
#Deseriallization:

import json
with open('demo.json','r')as f:
    a=json.load(f)  # no need to type converstion
    print(a)
    print(type(a))
```

```python
# Seriallization & Deseriallization tuple:
import json
t=(1,2,3,4,5)       # tuple store like list
with open('demo.json','w')as f:
    json.dump(t,f)
```

```python
# Seriallization & Deseriallization a nested dict:

d={
    'std':'Awais',
    'marks':[20,30,40,50]        # list inside a dict..
}
with open('demo.json','w')as f:
    json.dump(d,f)
```

 Seriallization & Deseriallization custom objects:

```python

class Person():
    def __init__(self,fname,lname,age,gender):
        self.fname=fname
        self.lname=lname
        self.age=age
        self.gender=gender
        
person=Person('Awais','Anas','21','male')
```

```python

import json

def show_obj(person):
    if isinstance(person,Person):            #If you need to check the type of an object use isinstance..
        return '{} {} age -> {} gender ->{}'.format(person.fname,person.lname,person.age,person.gender)
with open('demo.json','w')as f:
    json.dump(person,f,default=show_obj) 
```

```python
# as dict:

import json

def show_obj(person):
    if isinstance(person,Person):
        return {'name':person.fname+' '+person.lname,'age':person.age,'gender':person.gender}
with open('demo.json','w')as f:
    json.dump(person,f,default=show_obj,indent=4) 
```

```python
#deseriallization:

import json
with open('demo.json','r')as f:
    d=json.load(f)
    print(d)
    print(type(d))
```

# pickling:

```python
# pickling is the process whereby a python obj hierarchy is converted into a byte stream,and
# unpickling is the inverse operation, whereby a byte stream (from a binary file or byte-like obj) is 
# converted back into an obj hierarchy.
```

```python
#
class Person():
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def display_info(self):
        print("Hi my name is ",self.name,'and i am',self.age,'Years old')
        
p=Person('Awais',21)
```

```python
import pickle
with open("person.pkl",'wb')as f:
    pickle.dump(p,f)
```

```python
import pickle
with open("person.pkl",'rb')as f:
    p=pickle.load(f)
p.display_info()
```
