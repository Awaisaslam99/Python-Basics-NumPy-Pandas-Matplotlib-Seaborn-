Write a program that will tell the number of dogs and chicken are
there when the user will provide the value of total heads and legs.

```python
heads = int(input("Enter the total number of heads: "))
legs = int(input("Enter the total number of legs: "))

dogs = 0
chickens = 0

while chickens <=heads:
    dogs = heads - chickens
    calculated_legs = 4 * dogs + 2 * chickens
    
    if calculated_legs == legs:
        break 
    else:
        chickens += 1  

if chickens <= heads:
    print("Number of dogs:", dogs)
    print("Number of chickens:", chickens)
else:
    print("No valid solution exists for the given input.")
```

Write a function that accepts a list of strings and performs Bag of words
and convert it to numerical vectors.

```python
def bow(sentences):
    vocab=[]
    
    for i in sentences:
        vocab.extend(i.split())
        
        
    vocab=list(set(vocab))
    
    vector2d=[]
    
    for sentence in sentences:
        vector=[]
        
        for word in vocab:
            vector.append(sentence.count(word))
        vector2d.append(vector)
    return vector2d
sentences= [
    'Hello how are you',
    'I was waiting for your call',
    'where do you work',
    'lets meet someday',
    'hope you are fine'
]

bow(sentences)
```

Write a program to swap the key value pair for max and min values
Eg if the dict is like this {‘a’:1,’b’:2,’c’:3}
Output should be {a:3,b:2,c:1}


```python
d={'a':1,'b':2,'c':3}

max_val= max(d.values())
min_val= min(d.values())

for i in d:
    if d[i]==max_val:
        print(d[i])
        break
for j in d:
    if d[j]==min_val:
        print(d[j])
        break
d[i]=min_val
d[j]=max_val
print(d)
    
```

Write a program that accepts neighbors(set of 2D co-ordinates) and a
point(single 2D co-ordinate) and tells nearest neighbor(in terms of
euclidean distance)


```python
def nearst_n(n,you):
    def distance(c1,c2):
        return ((c1[0] - c2[0])**2 + (c1[1] - c2[1])**2)**0.5
    
    
    d=[]
    
    for i in n:
        d.append(distance(you,i))
        
    n_index, d= sorted(list(enumerate(d)),key=lambda x:x[1])[0]
    
    print('Neardt distancen is: ',n_index,'at the distance of: ',d)
    
n=[
    [89,34],
    [12,34],
    [10,89],
    [90,90]
]
you=[-9,7]
nearst_n(n,you)

```

Write a function that accepts a number and returns it’s factorial. You
can not use any loop

```python
def fact(num):
    if num==1:
        return 1
    else:
        return num*fact(num-1)
fact(5)
```
