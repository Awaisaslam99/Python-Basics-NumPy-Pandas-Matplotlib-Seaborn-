```python
import numpy as np
import pandas as pd
```

```python
courses= pd.read_csv('courses.csv')
student= pd.read_csv('students.csv')
nov= pd.read_csv('reg-month1.csv')
dec= pd.read_csv('reg-month2.csv')

matches= pd.read_csv('matches.csv')
delivery= pd.read_csv('deliveries.csv')


```

# Concatenation:

pd.concat:

```python
# it is use to concatenate two or more data-sets vertically... 

regs= pd.concat([nov,dec], ignore_index=True)--=-=
# you see in output indexes are repeat after 24, we can solve this issue by using parameter 'ignore_index'..

regs
```

append:

```python
# we can also concatenate datasets by using append..

nov.append(dec, ignore_index=True)  # suggestion: always use 'pd.concat' 
```

Multi-index Dataframe..

```python
# we can also set indexes of to diff datasets..see output ...
multi= pd.concat([nov,dec], keys=['Nov','Dec'])
# this is called multiindex Dataframe...



# we can also fetch only 'nov' or 'dec' dataset..
multi.loc['Nov'] # when we fetch by labels use loc...




# we can also fetch rows of datasets.. provide value in tuple...
multi.loc[('Nov',0)]  # etc
```

 previous we just concatenate dataset vertically, now we can also concatenate datasets horizently... 

```python
pd.concat([nov,dec],axis=1)  # dataset horizently
```

# Joins:

```python
#pandas joins in python are used to combine two dataFrames...

# There are 4 joins:

#1- Inner Join
#2- Outer Join also called Full Join...
#3- Left  Join
#4- Right Join
```

Inner Join:

```python
# Inner join in pandas is used to merge two data frames at the intersection. It returns the mutual rows of both...


# 1st table is left table and 2nd is right table...


# perform inner join on 'student' and 'regs' table...

# we want to inner join 'student' and 'regs' table...
student.merge(regs,how='inner',on='student_id')
# in this line of code 'merge' is a func of joins and regs is a 2nd table and in 'how' we tell which join 
# we want to perform and in 'inner join' we tell mutual cols in both table..'student_id' is mutual in both, so we tell this...



# result ma hummy only wo values hi mily gi jo dono table ma ha, e.g. 'student_id' dono table ma ha, but regs ma 42,50,38  
# aesi id ha jo 'student' ma nai ha, so humy result ma wohi mily gi jo dono ma hu..
```

left join:

```python
# A left join is also called Left Outer Join which returns all rows from the left DataFrame regardless
# of match found on the right DataFrame...



#perform left join on 'courses' and 'regs' table...
courses.merge(regs,how='left', on='course_id')
```

right join:

```python
#all the rows of the right dataframe are taken as it is and only those of the left dataframe that are common in both...

temp= pd.DataFrame({
    'student_id': [26,27,28],
    'name': ['Awais','Anas','junaid'],
    'partner': [28,26,27]
})
student= pd.concat([student,temp],ignore_index=True)  # we add some rows in stuent table..
student
```

```python
student.merge(regs,how='right', on='student_id')
```

Outer join:

```python
# it show all data present in both table..

# An outer join is a type of join operation that includes all the rows from both dataframes,
# matching them where possible based on a specified column or index. If there is no match for a row in one of the 
# dataframes, the values for the missing row will be filled with null or NaN values.


student.merge(regs,how='outer',on='student_id')
```

# Some Questions on joins:

```python
#1.find total revenue genrated...

regs.merge(courses,how='inner',on='course_id')['price'].sum()
```

```python
# 2. find month by month revenue....

temp_df= pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()

temp_df.merge(courses,on='course_id').groupby('level_0')['price'].sum() # inner join is default so we does not need to pass inner join...
```

```python
# 3. Print the registration table
# cols -> name -> course_name -> price

regs.merge(student,on='student_id').merge(courses,on='course_id')[['name','course_name','price']]
```

```python
# 4. Plot bar chart for revenue/course

regs.merge(courses,on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
# 5. find students who enrolled in both the months...

nov['student_id'],dec['student_id']  # this is a series so we can apply numpy operation...

common= np.intersect1d(nov['student_id'],dec['student_id'])
# ye numpy func dono ma jo common id's hu gi unky index nikal ly ga..
student[student['student_id'].isin(common)]
```

```python
# 6. find course that got no enrollment
#courses['course_id']
#regs['course_id']
                       # 'setdiff1d'  (A - B) operation, means wo nikal k dy ga jo A ma ha or B ma nai....
id_list= np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(id_list)]
```

```python
# 7. find students who did not enroll into any courses....
id_list= np.setdiff1d(student['student_id'],regs['student_id'])
student[student['student_id'].isin(id_list)]
```

```python
# 8. Print student name -> partner name for all enrolled students
# self join: joining the same table is called self join...same mean khud ko khud k saath join..
# self join is a variant of inner join..

student.merge(student,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']] #name_x is std_name, name_y is partner name..

# use 'left_on' and 'right_on' to join diff col...
```

```python
# 9. find top 3 students who did most number enrollments:

regs.merge(student,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# 10. find top 3 students who spent most amount of money on courses.....
regs.merge(student,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
# Alternate syntax for merge:


# previous we use this syntax..
#student.merge(regs)


# one more syntax..
pd.merge(student,regs, how='inner',on='student_id')
```
