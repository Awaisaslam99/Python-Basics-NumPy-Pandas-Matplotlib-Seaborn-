# Aggregation:

```python
# Aggregation (also known as **containment) is a specialised form of association. 
# It defines a one-way relationship that specifies a 'has-a' relationship between two classes. 
# For example, you can say that a wallet has money. 
# However, the money does not need a wallet in order to exist, so the relationship is one-way.
# e.g: Customer has a address, Resturant has a menu...etc
```

```python
# Example:
class Customer():
    def __init__(self,name,gender,address):
        self.name=name
        self.gender=gender
        self.address=address          # Here address is a multiple things so we can creat a new class..
        
    def print_address(self):
        print(self.address.city,self.address.pin,self.address.state)

class Address():
    def __init__(self,city,pin,state):
        self.city=city
        self.pin=pin
        self.state=state

A=Address('MDK',1234,'LHR')
C=Customer('Awais','male',A)

C.print_address()
```

```python
# Example with private variable with aggregation:
# Example:
class Customer():
    def __init__(self,name,gender,address):
        self.name=name
        self.gender=gender
        self.address=address          # Here address is a multiple things so we can creat a new class..
        
    def print_address(self):
        print(self.address.__city,self.address.pin,self.address.state)
                          #private  
class Address():
    def __init__(self,city,pin,state):
        self.__city=city    #private
        self.pin=pin
        self.state=state

A=Address('MDK',1234,'LHR')
C=Customer('Awais','male',A)

C.print_address()
# we can't access the owner class variable in aggregation.
# In aggrigation we can't access private attribute if you want to access see below....here we access other class attribute.
```

```python
# Now how can we access private attributes in aggregation ??
# by using get funtion we can access private attribute in aggregation.
# Example with private variable with aggregation:
# Example:
class Customer():
    def __init__(self,name,gender,address):
        self.name=name
        self.gender=gender
        self.address=address          # Here address is a multiple things so we can creat a new class..
        
    def print_address(self):
        
        # after get function use this syntax (self.address.get_city())
        print(self.address.get_city(),self.address.pin,self.address.state)
                          #private  
class Address():
    def __init__(self,city,pin,state):
        self.__city=city    #private
        self.pin=pin
        self.state=state
        
    def get_city(self):          # Get function
        return self.__city
        

A=Address('MDK',1234,'LHR')
C=Customer('Awais','male',A)

C.print_address()

 
```

# Inheritance:

```python
# Inheritance is the procedure in which one class inherits the attributes and methods of another class.
# The class whose properties and methods are inherited is known as the parent class or superclass. 
# And the class that inherits the properties from the parent class is the child class or derived class.
```

```python
# Example:

# Parent class:
class User():
    def __init__(self):
       self.name='Awais'
        
    def login(self):
        print("login")
        
# Child class:
class Student(User):

# If child has it's own constructor then parent constructor never call. And the things inside parent constructor never print.
    #  def __init__(self):
     #    self.rollno=100
        
    def enroll(self):
        print('Enroll into course')

        
u=User()
s=Student()
print(s.name)

s.login()
s.enroll()
```

```python
# Child can't access private  member of parent class.
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera
        
    def __show(self):
        print(self.__price)
        
class Smartphone(Phone):
    def check(self):
        print(self.__price)

s=Smartphone(20000,'Apple',13)
print(s.brand)  # we can access non-private memder.
#s.price()       # but can't access private member.

print(s.show())  # we can't access private Method.
```

```python
# We can access private attribute using get function:

class Parent():
    
    def __init__(self,num):
        self.__num=num
        
    # get function:
    def get_num(self):
            return self.__num
            
class Child(Parent):
    
    def show(self):
        print("This is child class")
        

c=Child(100)
print(c.get_num())
c.show()
```

Method Overriding:

```python
# # parent & child has same name Method, then it always print child method. This method is called Method Overriding.
```

```python
# Method overriding Exa
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera
        
    def buy(self):
        print("Buying a phone")
        
class Smartphone(Phone):
    # parent & child has same name Method, then always print child method. 
    # if you want to print both use super keyword.
    def buy(self):
        print("Buying a Smartphone")

s=Smartphone(20000,'Apple',13)
s.buy()
```

Super Keyword:

```python
# super keyword is a way to get parent method
# 1- Super can'tm access variable.
# 2- super can't be used outside the class.
# 3- Super is used inside the child class.
```

```python
# There is 3 keywords,
# 1- self
# 2- Static
# 3- super
```

```python
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera
        
    def buy(self):
        print("Buying a phone")
        
class Smartphone(Phone):
    
    def buy(self):
        print("Buying a Smartphone")
        # syntax to call parent buy method
        super().buy()

s=Smartphone(20000,'Apple',13)
s.buy()
```

```python
#Super constructor:
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera
        
        
class Smartphone(Phone):
    def __init__(self,price,brand,camera,os,ram):
        print("Inside smartphone constructor")
        
        #super keyword
        super().__init__(price,brand,camera)
        self.os=os
        self.ram=ram
        print("Inside Smartphone constructor")

s=Smartphone(20000,'Apple',12,"IOS",2)
print(s.os)
print(s.ram)
```

```python
# Super outside the class:
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera=camera
        
    def buy(self):
        print("Buying a phone")
        
class Smartphone(Phone):
    
    def buy(self):
        print("Buying a Smartphone")
        # syntax to call parent buy method
        super().buy()

s=Smartphone(20000,'Apple',13)
# syntax to call parent buy method
super().buy()

# We can't use super keyword outside the class, always use super inside the class.
```

# Inheritance Summary:

```python
# 1- A class can inherit from another class.
# 2- Inheritance improve code reuseability.
# 3- constructor,attributes,methods get inherited to the child class.
# 4- The parent has no access to the child class.
# 5- Private properties of parent are not accessible directly in child class.
# 6- Child class can override the attributes or methods. This is called method overriding.
# 7- super() is an inbuilt function which is used to invoke the parent class methods and constructor.
```

# Types of inheritance:

```python
# 1- single
# 2- Multilevel
# 3- Hierarchical (1 parent multiple child)
# 4- Multiple     (Multi parent one child)
# 4- Hybrid       (combination of inheritance)
```

```python
#Multilevel:
class Product():
    def review(self):
        print("Customer review")
class Phone(Product):
    
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.price=price
        self.brand=brand
        self.camera=camera
        
    def buy(self):
        print("Buying a phone")
        
class Smartphone(Phone):
  
    pass

s=Smartphone(20000,'Apple',13)
s.buy()
s.review()
```

```python
# Hierarical:
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.price=price
        self.brand=brand
        self.camera=camera
        
    def buy(self):
        print("Buying a phone")
        
class Smartphone(Phone):
   
    pass
class Featurephone(Phone):
    pass
Smartphone(20000,'Apple',13).buy()
Featurephone(10000,'OPPO',15).buy()
```

```python
#Multiple:
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.price=price
        self.brand=brand
        self.camera=camera
        
    def buy(self):
        print("Buying a phone")
        
class Product():
    def review(self):
        print("Customer review")
    
class Smartphone(Phone,Product):
    pass

s=Smartphone(20000,'Apple',13)
s.buy()
s.review()
```

```python
# Method overloading:
class Shape():
    def area(self,a,b=0):
        if b==0:
            return 3.14*a*a
        else:
            return a*b
    
a=Shape()
print(a.area(5))
print(a.area(2,3))
```

# Abstraction:

```python
# Abstraction is the process of hiding the internal details of an application from the outer world. 
# Abstraction is used to describe things in simple terms.
# It's used to create a boundary between the application and the client programs.
```

```python
# Example:

from abc import ABC, abstractmethod
class Bankapp(ABC):     #To define abstract class use ABC in class
    def database(self):
        print('Connected to database')
        
    @abstractmethod
    def security(self):
        pass
    
class Mobileapp(Bankapp):
    
    def log_in(self):
        print('Login to Mobile')
        
        # we can't inherit Bankapp class untill we use abstractmethod in child class.
    def security(self):
        print('Mobile security') 
        
a=Mobileapp()      #we can't make abstract class obj.
a.database()
```
