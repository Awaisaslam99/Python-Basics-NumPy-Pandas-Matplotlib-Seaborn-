```python
# Catching specific exception:


try:
    m=5
    f= open('sample.txt','r')  
    print(f.read())
    print(m) 
    print(5/2)
    L=[1,2,3]
    L[100]

# if we know what kind of error can happen in our prgrame...use this 'except'.....
except FileNotFoundError:              # This is a way to catch different type of errorss.......
    print('File not found')
except NameError:
    print("Variable not found")
except ZeroDivisionError:
    print("Can't divide by 0")
    
    
    # if you don't know what kind of can happpend during execution,,use 'generis exception'.....
except Exception as e:                # This is a genraic exception..
    print(e)
# use generic exception in the end of the code...

```

else:

```python
# There is not only 'try except' block, there is more block like 'else' :

try: # There is a two way in a 'try':
     # 1- if something wrong in the code it goes to 'except'..
     # 2- if ther is no prblem in the code it goes to 'else'...
    f= open('sample.txt','r')     
    
    # if there is something problem in the code then any 'except' can handle code...
except FileNotFoundError:
    print('File not found')
except Exception as e:
    print(e)
    
    
    # if there is no problem in the code then it goes to 'else'..
else:
    print(f.read())
    

# when you are sure the code inside the 'try' work properly, then use 'else'....
```

```python
try: 
    # 1- try
    # 2- if there is something problem in the code then it goes to 'except'...
    # 3- if there is no problem in the code then it goes to 'else'..
    # 4- if there is problem or no problem, it definitely goes to 'finally'....
    f= open('sample1.txt','r')     
    
    # if there is something problem in the code then any 'except' can handle code...
except FileNotFoundError:
    print('File not found')
except Exception as e:
    print(e)
    
    
    # if there is no problem in the code then it goes to 'else'..
else:
    print(f.read())
    
    
    
# kuch bhi hu, we must go to the 'finally'...
finally:
    print('This will definitely work')
    
# why we use finally:
# to execute code....
# 1- There is a chance that code can not complete it's execution, so 'finally' help to reach end of the code.
```

```python
class Bank():
    def __init__(self,balance):
        self.balance=balance
        
    def withdraw(self,amount):
        if amount<0:
            raise Exception("amount can't be -ve")
        if self.balance<amount:
            raise Exception("There is no money")
        self.balance=self.balance - amount
        
a=Bank(10000)

try:
    a.withdraw(15000)
except Exception as e:
    print(e)
    
else:
    print(a.balance)
```

Custom Exception classes

```python


class MyException(Exception):           #Custom Exception class
    def __init__(self,message):
        print(message)

class Bank():
    def __init__(self,balance):
        self.balance=balance
        
    def withdraw(self,amount):
        if amount<0:
            raise MyException("amount can't be -ve")
        if self.balance<amount:
            raise MyException("There is no money")
        self.balance=self.balance - amount
        
a=Bank(10000)

try:
    a.withdraw(15000)
except MyException as e:
    pass
    
else:
    print(a.balance)
```
