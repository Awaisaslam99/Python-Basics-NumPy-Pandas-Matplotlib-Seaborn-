# List Comprehension:

```python
# list comprehension provide a consise way of creating lists.
#   syntax :  new list= [ expression for i in iterable if condition==True]
```

```python
#add 1 to 10 numbers in list:
L=[]

for i in range(1,11):
    L.append(i)
print(L)
```

```python
#same code using comprehension:
L=[i for i in range(1,11)]
print(L)
```

```python
# Scalar multiplication on a vector:
v=[2,3,4]
s=-3
x=[]
for i in v:
    x.append(i*s)
    
print(x)
```

```python
# using comprehension:
[s*i for i in v]
```

```python
# add squares:
L=[1,2,3,4,5]
[i**2 for i in L]              # using comprehension:
```

```python
# print all numbers divicible by 5 in the range of 1 to 50:
[i for i in range(1,51) if i%5==0]                # using comprehension:
```

```python
# find languanges which start with letter p:
languages=['java','python','php','javascript']
[language for language in languages if language.startswith('p')]        # using comprehension:
```

```python
# Nested if with list comprehension:
basket=['apple','guava','cherry','banana']
my_fruits=['apple','kiwi','grapes','banana']
# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a':
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# print (3,3) matrix :
[i*j for i in range(1,4) for j in range(1,4)]
```

```python
# cartesian product:
l1=[1,2,3,4]
l2=[5,6,7,8]
[i*j for i in l1 for j in l2]
```
