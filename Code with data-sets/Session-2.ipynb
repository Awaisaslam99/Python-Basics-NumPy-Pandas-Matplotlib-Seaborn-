```python
# While loop:
u1=int(input('Enter Number'))
i= 1

while i<11:
    print(u1 *i)
    i +=1
```

```python
# while loop with else:
x=1
while x<3:
    print(x)
    x+=1
else:
    print('Limit crossed')
```

```python
# Guessing Game:
import random
j=random.randint(1,100)


guess=int(input('Guess Number'))
counter=1
while guess!=j:
    if guess<j:
        print('Wrong, Guess higher')
    else:
        print('Wrong, Guess lower')
    
    guess=int(input('Guess Number'))
    counter +=1
else:
    print('Correct Guess')
    print("Attempts", counter)
```

```python
#print pattern:
rows=int(input('Enter no.'))

for i in range(1,rows+1):
    
    for j in range(1,i+1):
        print('*',end='')
    print()
```

```python
# one more pattern example:
rows=int(input('Enter no.'))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end='')
    for k in range(i-1,0,-1):
        print(k,end='')
        
        
    print()
```

#  Loop control statement:

Break,   
continue,   
pass

```python
#Break statement:
for i in range(1,10):
    if i==5:
        break
        
    print(i)

```

```python
#Break statement: program:

lower=int(input('Enter lower range :'))
upper=int(input('Enter upper range :'))

for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i)
```

```python
#Continue Statement:
for i in range(1,10):
    if i==5:                  #skip this number
        continue
    print(i)
```

```python
#pass statement:
for i in range(1,10):
    pass
    
        
    
```
