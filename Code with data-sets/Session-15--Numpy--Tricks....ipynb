# 1.np.sort

```python
# Return a sorted copy of an array..

import numpy as np
 
    # 1D
a= np.random.randint(1,100,15)
print(a)

np.sort(a)
```

```python
# 2D
b= np.random.randint(1,100,24).reshape(6,4)
print(b)

print(np.sort(b))  # sort row vise..
print(np.sort(b,axis=0))   # sort column vise
```

# 3.np.concatenate:

```python
# numpy.concatenate() func concatenate a sequence of array along an existing axis..

c= np.arange(6).reshape(2,3)
d= np.arange(6,12).reshape(2,3)
print(c)
print(d)

# we want to concatenate both c,d..
print(np.concatenate((c,d),axis=0))   # concatenate in column
print(np.concatenate((c,d),axis=1))   # concatenate in row
```

# 6.np.where:

```python
# The np.where() func return the indices of elements in an input array where the given condition is satisfied.

print(a)

print(np.where(a>50))  # find all indices with value greater than 50..
```

# Set Functions:

1.np.union1d          
2.np.intersect1d          
3.np.setdiff1d        
4.np.setxor1d        
5.np.in1d       

```python
m= np.array([1,2,3,4,5])
n= np.array([3,4,5,6,7])


print(np.union1d(m,n))  # it calculate union..

print(np.intersect1d(m,n))  # tell intersection..


#setdiff1d() function find the set difference of two arrays and return the unique values in arr1 that are not in arr2..
print(np.setdiff1d(m,n)) 


# setxor1d() function find the set exclusive-or of two arrays and return the sorted, unique values that are in
# only one (not both) of the input arrays.
print(np.setxor1d(m,n)) 


# in1d() function test whether each element of a 1-D array is also present in a second array and return a
# boolean array the same length as arr1 that is True where an element of arr1 is in arr2 and False otherwise.
print(np.in1d(m,1))
```
