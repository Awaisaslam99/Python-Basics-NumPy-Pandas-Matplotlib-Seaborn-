*args and ***kwargs:

```python
# *args &  **kwargs:
# *args & **kwargs are special python keywords that are use to pass the variable lenght of arguments to a function.
```

```python
# *args:
# allow us to pass a vaiable number of a non-keyword argument to a function.
# use *args when you don't know that how much numbers user enter..
def multi(*args):     #use anything instead of args like  *a, *b, *c, etc...
    product=1
    
    for i in args:
        product = product *i
    print(args)
    return product
multi(1,2,3,4,4,5,5,6,7)
```

```python
#**kwargs:
# **kwargs allow us to pass any number of keyword argument.
# Keyword argument means that they contain  a key-value pair, like a python dictionary.

def display(**kwargs):                 #use anything instead of kwargs like  *a, *b, *c, etc...
    for (key,value) in kwargs.items():
        print(key,'->',value)
        
display(pakistan='Islamabad',India='Delhi',srilanka='Colombo')
    
```

```python
#Points to remember while using *args &  **kwargs:
# 1- order of the arguments matter!(normal-> * args-> **kwargs).
# 2- The word 'args' and 'kwargs' are only a convention, you can use any name of your choice.
```

# Lambda Function:

```python
# A lambda function is a small anonymous function.   {anonymous "means func without name"}
# A lambda function can take any number of arguments, but can only have one expression.

```

lambda keyword:

```python

#lambda a,d:a+b
#here a,b are parameters & a+b is a expression..
```

Creating lambda function:

```python
# x-> x^2:
a=lambda x:x**2   # to run lambda function then store it in a variable..
a(2)
```

```python
# x,y->x+y:
b=lambda x,y:x+y
b(2,3)
```

Diff b/w lambda and normal function:

```python
# 1- No name
# 2- lambda has no return value(infact, return a function)
# 3- lambda is written in 1 line
# 4- not reusable
```

Then why we use lambda functions???

```python
# They are used with Higher order function(HOF).
```

```python
# just creating lambda function:
# check if a string has 'a' :
c= lambda s:'a' in s 
c('hello')
```

```python
# just creating lambda function:
# odd or even:
d=lambda x:'even' if x%2==0 else 'odd'
d(5)
```

Higher Order Function(HOF):

```python
# The function that return a function is know as HOF.
```

```python
#Example:
def square(x):
    return x**2

#HOF                                      1:59:52
def transform(f,L):
    output= []
    for i in L:
        output.append(f(i))
    print(output)
    
L=[1,2,3,4,5]
transform(lambda x:x**4 ,L)
```

Map:

```python
# Map must have two inputs....
```

```python
# square the items of list:
list(map(lambda x:x**2,[1,2,3,4,5]))
```

```python
# odd/even labeling of list:
L=[1,2,3,4,5]
list(map(lambda x:'even' if x%2==0 else 'odd' ,L))
```

```python
# Fetch names from a list of dict:
users=[
    {
        'name':'Awais',
        'Age':21,
        'Gender':'Male'
    },
    {
        'name':'Mahad',
        'Age':23,
        'Gender':'Male'
    },
    {
        'name':'Umair',
        'Age':22,
        'Gender':'Male'
    }
]

list(map(lambda users:users['name'],users))
```

Filter:

```python
#numbers greater than 5:
L=[3,4,5,6,7,8]
list(filter(lambda x:x>5,L))
```

```python
# fetch fruits with 'a' :
f=['apple','guava','cherry']
list(filter(lambda x:x.startswith('a'),f))
```

Reduce:

```python
# sum of all items:
import functools
functools.reduce(lambda x,y:x+y,[1,2,3,4,5])
```

```python
# find min:
L=[1,2,3,4,5]
functools.reduce(lambda x,y:x if x<y else y,L)
```
