Categorical Scatter Plot:

```python
# we use this plot when we want to see the distribution of categorical col with respect to a numerical col..
# they are a bivariant..
```

```python
sns.scatterplot(data=tips, x='total_bill',y='tip') # this is scatter plot on numerical data..

# we can alsp creat scatter plot on categorical data...
# but u have to make sure your 1st-col is numerical & 2nd-col is categorical..
```

```python
# strip plot:is use for visualizing the distribution of a continuous variable within each category of a categorical variable.

# axes level function

sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
# see in output the dosts not in one line, to make it in one line, use 'jitter=False'...
```

```python
# using catplot
# figure level function

sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
# jitter:

sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex') # you can control jitter size...
```

```python
# swarmplot:is a function that plots a scatterplot of the data with the points spread out along the x-axis to avoid overlap


# figure level..
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
# axes level..
sns.swarmplot(data=tips,x='day',y='total_bill')
```

```python
# hue:
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

Categorical Distribution Plots:

```python
# it tells the distribution of the data..
# and they are univariant...
```

```python
# Boxplot:
# A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, 
# first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their 
# values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how 
# your data is skewed.....

sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
# Using catplot:


sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
# single boxplot -> numerical col:

sns.boxplot(data=tips,y='total_bill')
```

Violinplot = (Boxplot + KDEplot)

```python
# violinplot:

sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
# hue:

sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
# barplot:

import numpy as np
# avg total bill of male, female...                                          
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min,ci=None) # to remove error bar use'ci=None'..
# we can also apply numpy operations...
```


When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval 
around the estimate, which is plotted using error bars



A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
# point plot:

sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)

# in this graph we connect male bar with male, and female bar with female...this is point plot..
```

```python
# countplot: it's like 'value_counts' func.. it count the values..

sns.countplot(data=tips,x='sex',hue='day')  # it count the male, female come on diff days...
```

```python
# faceting using catplot:

sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',kind='box',row='time')
# default value is strip plot, but we can change it..
```

# 5-Regression Plots:

```python
# regplot:creates a regression line between 2 parameters and helps to visualize their linear relationships..
# lmplot
```

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# regplot:
# axes level:
# hue parameter is not available....

sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')  # hue is the diff between reg & lmplot..
```

```python
# residplot: is used to plot residual data of a linear regression...

sns.residplot(data=tips,x='total_bill',y='tip')
```

# A second way to plot Facet plots -> FacetGrid

```python
# figure level -> relplot -> displot -> catplot -> lmplot

sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
# there is 1st way to creat facet-plot, jo humne prh chuky ha...
```

```python
g= sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.violinplot,'sex','total_bill')
```

#  Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
sns.pairplot(iris,hue='species')
```

```python
# pair grid:

g= sns.PairGrid(data=iris,hue='species')

# g.map

g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag:

g= sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)   

# you can also change plots..in map_diag -> map_offdiag...
```

```python
# map_diag -> map_upper -> map_lower

g= sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars


g= sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

# JointGrid Vs Jointplot

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
g= sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
