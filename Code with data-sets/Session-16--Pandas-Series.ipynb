# Series from lists:

```python
# strings:

country= ['Pak','India','Nepal','Afg','USA']

pd.Series(country)  # it creat series...


# number in the output is called index of the item..
```

```python
# integers:

runs= [13,24,64,67,50,56]

run_ser=pd.Series(runs)
run_ser
```

```python
# custom index:

# in previos examples index are automatically generate..we can also generate index by using "index" ...

marks=[67,47,87,45]

subjects= ['math','english','urdu','Science']

pd.Series(marks, index= subjects)
```

```python
# Setting a name:
# we can also give a name to our series by using "name"...

result=pd.Series(marks, index= subjects, name='Awais Result')
result
```

10- copy:

```python
# when we use head & tail tu ye hummy original data deta ha na k copy , agr hum head & tail ko edit 
# kry ga tu original data ma changing hu jay ga, so is se bachny k ley hum "copy" func use krty ha,
# "copy" func hummy copy dy deta ha or jb copy ma change krty ha tu original waly ma koi change nai hota ha..

new=vk.head().copy()  # it  will creat copy..

new[1]= 100  #  it change the 1 value into 100..but it will not change original data, 
new


# but head & tail can change the oiginal data..
```

```python
# Write a function to change the data type of given a column or a Series. 
# Function takes series and data type as input, returns the converted series...

def change_data_type(series,dtype):
    return pd.to_numeric(series, errors='coerce', downcast= dtype)

pd.Series([1,2,'Python', 2.0, True, 100])
change_data_type(pd.Series([1,2,'Python', 2.0, True, 100]),'float')
```
