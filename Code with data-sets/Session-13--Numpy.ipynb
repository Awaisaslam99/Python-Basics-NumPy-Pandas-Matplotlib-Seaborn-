Numpy Arrays vs python sequences:

```python
#1- Numpy arrays have a fixed size at creation,unlike python lists(which can grow dynamically). Changing the size of 
# an ndarray will creat a new array and delete the original.

#2- The elements in a numpy array are all required to be of the same data type, and thus will be the same size in memory.

#3- Numpy arrays facilitate advanced mathematical and other types of opeartions on large number of data. Typically such
# operations are executed more efficiently and with less code than is possible using python's built-in sequences.

#4- A growing plethora of scientific and mathematical python-based packages are using numPy arrays; through these typically
# support python sequence input, they convert such input to numPy arrays prior to processing, and they often output numPy
# arrays
```

# Arrays Functions:

```python
a1= np.random.random((3,3))
a1= np.round(a1*100)  # round use to roundoff the items..
a1
```

```python
# max  : it tells max n umber in matrics...
print(a1)
print("Max: ",np.max(a1))

#min
print('Min: ',np.min(a1))

# sum
print('Sum: ',np.sum(a1))

# product: give mulitplications...
print('Product: ',np.prod(a1))
```

# indexing & slicing:

```python
a1= np.arange(10)
a2= np.arange(12).reshape(3,4)
a3= np.arange(8).reshape(2,2,2)
```

```python
# positive 7 nagetive indexing:

#1D
print(a1)
a1[0]
a1[-1]

```

```python
# 2D indexing:
print(a2) # we need to find 6..

# 1st tell row and then tell col..three row in a2 "0,1,2"
a2[1,2]
```

```python
# 3D indexing:
print(a3)  # find 5..
# in a3 there is two  2D in 3D "0,1", so 1st tell kis 2D ma number ha 5 1st 2D ma ha, and then tell row and then col...
print(a3[1,0,1])  # find 5
print(a3[0,1,1]) # find 3
print(a3[0,0,0]) # find 0
```

Slicing:

```python
# 1D
print(a1)
a1[2:5]
```

```python
# 2D
print(a2)

print(a2[0,:])  #1st tell row and then col...we need all col so we use ":"..

print(a2[:,2])    
```

```python
# 2D
print(a2)
print()
print(a2[1:,1:3])   # 5,6,9,10
print()
print(a2[0::2,0::3])   # 0,3,8,11
print()
print(a2[0::2,1::2])  # 1,3,9,11
print()
print(a2[1,0::3])  # 4,7
print()
print(a2[0:2,1:])    # 1,2,3,5,6,7
print()
print(a2[1:,0::2])   # 4,6,8,10
```

```python
# 3D
a3= np.arange(27).reshape(3,3,3)
print(a3)
print("*************************")

#print(a3[1,1,:])   # bech waly ka 2nd row
print()
#print(a3[2,1:,1::])   # 22,23,25,26
print()
print(a3[::2,0::3,0::2])   # 0,2,18,20
print()
print(a3[::2,2::2,0::2])       # 6,8,21,23
```

# Stacking:

```python
# Stacking is the concept of joining arrays in NumPy. Arrays having the same dimensions can be stacked.
```

```python

a4= np.arange(12).reshape(3,4)
a5= np.arange(12,24).reshape(3,4)
a4
```

```python
# horizontal stacking:

np.hstack((a4,a5))
```

```python
# vertical stacking:
print(a4)
print(a5)
np.vstack((a4,a5))
```
