```python
#Example.. ATM code:

class ATM():
    #Note:
    # 1- If you want to make variable inside class creat variable inside the constructor.
    # 2- use self. before the variable name.
    
    # Constructor: Special function.constructor call itself.
    def __init__(self):   # self is the current object..
        self.pin=''
        self.balance= 0
        # call menu function.
        self.menu()
        
    def menu(self):
        user_input= input("""
        Hi! How can i help you ?
        
        1: Enter 1 to create key.
        2: Enter 2 to change pin.
        3: Enter 3 to check balance.
        4: Enter 4 to withdraw.
        5: Enter anything to exit.
    
        """)
        
        if user_input=='1':
             self.creat_pin()
        elif user_input=='2':
             self.change_pin()
        elif user_input=='3':
             self.check_balance()
        elif user_input=='4':
             self.withdraw()
        else:
            exit()
            print('Have a good day!')

    def creat_pin(self):
        user_input= input("Enter your pin")
        # use variable that creat in constructor.
        self.pin=user_input
        
        user_balance=int(input('Enter balance'))
        # use variable that creat in constructor.
        self.balance= user_balance
        
        print('pin created successfully!')
        self.menu()
    
    def change_pin(self):
        old_pin= input('Enter old pin')
        
        if old_pin==self.pin:
            new_pin=input('Enter new pin')
            self.pin=new_pin
            print('pin change successful')
            self.menu()
        else:
            print('Nope! thats wrong pin')
            self.menu()
            
    def check_balance(self):
        user_pin=input('Enter your pin')
        if user_pin==self.pin:
            print('Your balance is  ',self.balance)
            self.menu()
        else:
            print('Get-out')
            self.menu()
            
    def withdraw(self):
        user_pin=input('Enter your pin')
        if user_pin==self.pin:
            amount=int(input('Enter withdraw amount'))
            if amount<=self.balance:
                self.balance=self.balance - amount
                print('Withdrawl successfully. Balance is:  ', self.balance)
                
            else:
                print("Nikl bsdk")
                
        else:
            print('Get-out')
            
        self.menu() 
    
obj=ATM()
```

# Magic Methods / Dunder Methods:

```python
# is a special method. like constructor..

#  syntax:  __name__

# using magic method we can make our own data-types.
```

Constructor:

```python
# is a magic method. it call itself.
```

Make our own data-type name fraction:

```python
class Fraction():
    # Parameterized constructor
    def __init__(self,x,y):
        self.num=x
        self.den=y
        
        # usign 2nd magic method str.
    def __str__(self):
        return '{}/{}'.format(self.num,self.den)        
      
    def __add__(self,other): 
        new_num= self.num *other.den+self.den *other.num  
        new_den= self.den*other.den
        
        return '{}/{}'.format(new_num,new_den)
    def __sub__(self,other):
        
        new_num= self.num *other.den-self.den *other.num  
        new_den= self.den*other.den
        
        return '{}/{}'.format(new_num,new_den)
    def __mul__(self,other):
      
        new_num= self.num *other.den * self.den *other.num  
        new_den= self.den*other.den
        
        return '{}/{}'.format(new_num,new_den)
    def __truediv__(self,other):
        
        new_num= self.num *other.den 
        new_den= self.den*other.num
        
        return '{}/{}'.format(new_num,new_den)
        
        
        #you can also use non magic method.
    def convert_to_decimal(self):
        return self.num/self.den
    
fr1=Fraction(2,3)
fr2=Fraction(1,2)
print(fr1)
```

```python
fr1.convert_to_decimal()
```

```python
print(fr1+fr2)
print(fr1-fr2)
print(fr1*fr2)
print(fr1/fr2)
```
